python元组

元组用 () 标识。内部元素用逗号隔开。但是元组不能二次赋值，相当于只读列表。

```python
tuple = ('runoob',786,2.23,'john',70.2)
tinytuple = (123,'john')
print(tuple)
print(tinytuple)
print(f"tuple[0]: {tuple[0]}")
print(tuple[1:3])
```

1、若 x 为纯数字，则不能有 base 参数，否则报错；其作用为对入参 x 取整

```python
print(int(2.1, base=16)) #报错
```

```python
print(int(2.1))
```

2、若 x 为 str，则 base 可略可有。

```python
print(int("17", base=8))
```
